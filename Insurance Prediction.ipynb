## Model part

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(data)
data1=scaler.transform(data)
data1=pd.DataFrame(data1,index=data.index,columns=data.columns)
data1.head()
```

```python
data1=data1.drop(labels=["id","Driving_License","Previously_Insured","Gender","Yes","< 1 Year","> 2 Years"],axis=1)

data1["id"]=data["id"]
data1["Driving_License"]=data["Driving_License"]
data1["Previously_Insured"]=data["Previously_Insured"]
data1["Gender"]=data["Gender"]
data1["Yes"]=data["Yes"]
data1["< 1 Year"]=data["< 1 Year"]
data1["> 2 Years"]=data["> 2 Years"]
```

```python
#spliting data
train1=data1[:304887]
test1=data1[304887:]
```

```python
train1["Response"]=train["Response"]
```

```python
train1.drop(labels=["id"],inplace=True,axis=1)
```

```python
train1.shape
```

```python
test.head()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x=train1.drop(labels=["Response"],axis=1)
y=train1["Response"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
plt.hist(y_train)
plt.grid()
plt.hist(y_test)
plt.legend(["Train","test"])
plt.show()
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression( penalty='l2',
    dual=False,
    tol=0.0001,
    C=1,
    fit_intercept=True,
    intercept_scaling=1,
    class_weight=None,
    random_state=None,
    solver='lbfgs',
    max_iter=1000,
    multi_class='warn',
    verbose=0,
    warm_start=False,
    n_jobs=None,
    l1_ratio=None)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)

sum(y_pred==y_test)/len(y_test)
#0.8767424316966775
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix

#1 means good 0 means bad
f1_score(y_test,y_pred)  #test acc
#0.027432712215320912
```

```python
y_pred=model.predict(x_train)
f1_score(y_train,y_pred) #train acc
#0.0259723602783211
```

```python
#for test data
y_pred=model.predict(x_test)

confusion_matrix(y_test,y_pred)
#(tn, fp, fn, tp)
```
